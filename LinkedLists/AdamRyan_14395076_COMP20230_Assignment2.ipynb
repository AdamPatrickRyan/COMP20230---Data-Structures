During the creation of LinkedLists, Linked Nodes, and Stacks/Queues, I have created two sets of classes.

The first is the LinkedNode/LinkedList/Stack/Queue classes. In these classes I have defined linked equality to be such that two nodes are equal if they share the same references and values. In the methods for LinkedLists/Stacks/Queues, I have implemented a check such that the user cannot add two equal nodes to the start or end of the list. This is done so to protect these classes against mistaken repeated entries. However, I know that traditionally LinkedLists allow the the addition of consecuritve duplicate nodes.

Therefore, I have created Alt classes which changes the equality condition to be that two nodes are equal if they have the same Python ID. This allows the addition of nodes with the same values so long as the nodes are different objects. This condition is solely designed to prevent against duplicate user entry mistakenly into the LinkedList class. These Alt classes are otherwise functionally identical. These are not redundant classes but are instead alternative implementations of the ADTs with less stringent duplicate entry checks.

A traditional linkedlist can be generated by removing the equality check on these methods. This is done as there was not explicit outlines of what test cases or the expected behaviour is for these objects with regards to duplicates repeatedly being added but is a design decision which I have implemented.

Test cases passed by LinkedLists will apply also to AltLinkedLists.

### Task 1: Linked Node Class

We define a Linked Node which is the individual components of the Linked List.

1. We initialise it with two variables: The current value, and a pointer to the next node. Default values of both are None.

---

2. We make a getter and setter for the elements for both.

<b>Complexity:</b>
Trivially O(1).


---

3. We define appropriate Magic Methods for the Linked Node.
    1. String form: The Current Node and Whether the node is linked or not.
    2. Representation: The string form.
    3. Equality: Two nodes are equal if they have the same value and link to the same node.
    4. Length: Nodes have length of 1.
    5. Greater Than/Less Than/ Less Equal/ Greater Equal: This compares the value of nodes but not the pointer.
    

<b>Complexity:</b>
These are trivially O(1).

```python
class LinkedNode:
    """A class for a node within a linked list.

    Each node should contain a value and a pointer to the next node."""
    
    __current_node_value=None
    __next_node_pointer=None
    
    
    ###------------BEGIN: INITIALISER-----------------------###
    def __init__(self,value=None,pointer=None):
        """Initialise a Linked Node"""
        
        self.set_current_value(value)
        self.set_next_node(pointer)
        
    ###------------END: INITIALISER-----------------------###
    
    
    
    
    
    ###------------BEGIN: CURRENT VALUE METHODS------------###
      
    ###Current Value
    def set_current_value(self,node_value):
        """A setter for current node value"""
        
        self.__current_node_value=node_value
        
        return 
        
        
    def get_current_value(self):
        """A getter for current node value"""
        return self.__current_node_value
    
    
    ###------------END: CURRENT VALUE METHODS------------###
    
    
    
    
    
    ###------------BEGIN: NEXT NODE METHODS------------###
        
    ###Next Node
    def set_next_node(self,pointer=None):
        """A setter for next node"""
        
        point=None
        
        #Only allow linked nodes to point to linked nodes not the current node
        if isinstance(pointer,LinkedNode):
            
            #Not this node
            if id(pointer)!=id(self):
                point=pointer
            
        
        self.__next_node_pointer=point
        
        return
        
    def get_next_node(self):
        """A getter for the next node"""
        return self.__next_node_pointer
    
    ###------------END: NEXT NODE METHODS------------###
    
    
    
    
    
    ###------------BEGIN: MAGIC METHODS------------###
    
    ###String Method
    def __str__(self):
        """Set the string value"""
        return """Node Value: {} | Linked: {}""".format(self.get_current_value(),self.get_next_node() is not None)
    
    ###Representation Method - Note: Risk of Infinite Loop if add node ref: N1 - N2 - N3 - N1-...
    def __repr__(self):
        """Display representation"""
        
        return self.__str__()
    
    
    #####---COMPARISON METHODS---###
    
    ###Equality Method
    #####NOTE: Nodes are equal only if values are equal and both point to the same next object.
    def __eq__(self, other):
        
        #Curent Value is equal
        if self.get_current_value()==other.get_current_value():
            
            #Current Value is equal and both are linkednodes
            if self.get_current_value() is not None and other.get_current_value() is not None:
                     
                #Next Nodes are None in both and equal
                if self.get_next_node() is None:
                    if other.get_next_node() is None:
                        return True
                    else:
                        return False
                
                #One is None and the other isn't
                else:
                    if other.get_next_node() is None:
                        return False
                
                #Neither are none
                if self.get_next_node()==other.get_next_node():
                    return True
            
            #Both are none
            elif self.get_current_value() is None and other.get_current_value() is None:
                return True
            
            #One is none but the other isn't
            else:
                return False
                
        else:
            return False

        
    def __gt__(self, other):
        if type(self.get_current_value())==type(other.get_current_value()):
            if self.get_current_value()>other.get_current_value():
                return True
            else:
                return False
        else:
            return False
        
    def __ge__(self, other):
        if type(self.get_current_value())==type(other.get_current_value()):
            if self.get_current_value()>=other.get_current_value():
                return True
            else:
                return False
        else:
            return False
 
    def __lt__(self, other):
        if type(self.get_current_value())==type(other.get_current_value()):
            if self.get_current_value()<other.get_current_value():
                return True
            else:
                return False
        else:
            return False
        
    def __le__(self, other):
        if type(self.get_current_value())==type(other.get_current_value()):
            if self.get_current_value()<=other.get_current_value():
                return True
            else:
                return False
        else:
            return False
        
    #####---COMPARISON METHODS---###
        
    def __len__(self):
        """The length of a linked node is 1 as it's 1 node"""
        return 1
    
    
    
    ###------------END: MAGIC METHODS------------###
```

### Task 1: Linked List Class

We define a Linked List which is comprised of Linked Nodes.

---

1. We initialise it with one variables: The first value in the Linked List.

---

2. We make a getter and setter for the header element.

<b>Complexity:</b>
Trivially O(1).

---

3. add_first: 
    1. Get the value of the header
    2. If the new value being passed is not null, and the new value is not a linked node, we make it a linked node.
    3. If the header is None, the list is empty, so we call get setter method.
    4. If the header is not None, the list is non-empty. If the header is not already the element we are trying to add, we link the new node to the original header and set the header to the new node.
   
<b>Complexity:</b>
This will be O(1): Static number of checks (first element). Form is T(n)=a+k which is O(1).

---

4.  add_last
    1. Get the value of the header
    2. If the new value being passed is not null, and the new value is not a linked node, we make it a linked node.
    3. If the header is None, the list is empty, so we call get setter method.
    4. If the header is not None, the list is non-empty. While the link to the next node is not none, we continue to get the next node.
    5. Once the next node is None, we are at the tail of the linked list.
    6. If the tail is not the Node we are trying to add, we set the current tail to point to the element we are trying to add, otherwise the tail is already the element trying to add.
 
<b>Complexity:</b>
This will be O(n): Need to loop through the linked list of length n with my implementation. Linear factors (the checks, calling methods) are not variable with n hence it'll be of form T(n)=an+k which means T(n) is O(n).

---

5.  remove_first
    1. Get the value of the header
    2. If the header is none, then there is no first element so done.
    3. If the header is not none, we set the first element to be the element linked to the current first node.

<b>Complexity:</b>
This will be O(1): Static number of checks (first element) and assignments. Form is T(n)=a+k which is O(1).


---


5.  traverse_list
    1. Print the string representation of the Linked List
        1. If the header is a LinkedNode, then the List is non-empty. Otherwise, the list is empty so print that it's empty.
        2. For every node in the list, print the string representation of the node.
        3. When the pointer is None, we have reached the end of the list so stop printing.
   
   
<b>Complexity:</b>
This will be O(n): Need to loop through the linked list of length n with my implementation. Linear factors (the checks, calling methods, printing, etc.) are not variable with n hence it'll be of form T(n)=an+k which means T(n) is O(n).

---


6. We define appropriate Magic Methods for the LinkedList.
    1. String form: The elements of the Linked List and whether the nodes are linked or not. (this is O(n) as above)
    2. Representation: The string form. (this is O(n) as above)
    3. Equality: Two LinkedLists are equal if all nodes in the linked list are equal (values and pointers match) and the LinkedLists are of equal size. (this is O(n) as all elements must be checked)
    4. Length: The number of nodes in the linked list (this is O(n) as above).

```python
class LinkedList:
    """A class for a linked list. A linked list is a series of linked nodes.."""
    
    __header=None
    
    ##------------BEGIN: INITIALISER-----------------------###
    def __init__(self,header=None):
        """Initialise a Linked Node"""
        self.set_header_value(header)
        
    ##------------END: INITIALISER-----------------------###
    
    
    ##------------BEGIN: HEADER VAL METHODS-----------------------###
    
    ###Header
    def set_header_value(self,header_value=None):
        """A setter for current node value"""
        
        if not isinstance(header_value,LinkedNode) and header_value is not None:
            print("Warning: You did not enter a linked node. You entered {}. Converting to node".format(type(header_value)))
            header_value=LinkedNode(header_value)
            
        self.__header=header_value
            
        return
        
    def get_header_value(self):
        """A getter for current node value"""
        return self.__header
    
    ##------------END: HEADER VAL METHODS-----------------------###
    
        
    ##------------BEGIN: ASSIGNMENT METHODS-----------------------###
    def add_first(self,new_header_node=None):
        """A method to add a new first element to the linked list"""
        
        header=self.get_header_value()
        
        #Checked that it's a new linked node, otherwise you shoudn't call it.
        if not isinstance(new_header_node, LinkedNode) and new_header_node is not None:
            print("Warning: You did not enter a node. You entered {}. Turning this into a node.".format(type(new_header_node)))
            new_header_node=LinkedNode(new_header_node)
            
        #If the Header Value is none, then the linked list was not defined so there is no existing header
        if header is None:

            #Set the header value to be the new node and done
            self.set_header_value(new_header_node)

        #The header value was not none, so it is a node.
        else:

            #Not adding a duplicate header
            if new_header_node!=header:
                #Set the new node to point to the old header node.
                new_header_node.set_next_node(header)

                #Set the header to be the new header node.
                self.set_header_value(new_header_node)

        return
    
    def add_last(self, new_tail_node=None):
        """A method to add a new node to the end of the linked list"""
        
        check_node=self.get_header_value()
        
        #Checked that it's a new linked node, otherwise you shoudn't call it.
        if not isinstance(new_tail_node, LinkedNode) and new_tail_node is not None:
            print("Warning: You did not enter a node. You entered {}. Turning this into a node.".format(type(new_tail_node)))
            new_tail_node=LinkedNode(new_tail_node)
        
        
        #The header is not empty
        if check_node is not None:
            
            #The original Node
            original_node=check_node
            
            #Check if the next link is None
            check_node=original_node.get_next_node()

            #Go through each node until the next node is none
            while check_node is not None:
                
                #The next node is not None, so set the header as the next node
                original_node=check_node
                
                #Check if the next node is None.
                check_node=original_node.get_next_node()

            
            #The header is now the tail as the next Node now returns None
            original_node.set_next_node(new_tail_node)
            
            #Set the New Node to have no Link as its the tail
            new_tail_node.set_next_node()
            
            return
            
        #The header is empty to begin with. I.e. there is no existing nodes
        else:
            self.set_header_value(new_tail_node)
        
        return
    
    def remove_first(self):
        """A method to remove the first element in the linked list"""
        
        #Get the next node from the header. 
        #--If it's not linked it's None so you're setting list to none.
        #--If it is linked it's a Node, so you're setting header to next node.
        
        header=self.get_header_value()
        
        if header is not None:
            self.set_header_value(header.get_next_node())
            
        return
    
    def traverse_list(self):
        """A method to traverse the list and print all elements"""

        print(self)
        
        return
    
    ##------------END: ASSIGNMENT METHODS-----------------------###
    
    
    ##------------BEGIN: EXTRA METHODS-----------------------###
    
    def get_first(self):
        """A method to get the first element of the linked list
        
        NOTEL Equivalent to get_header_value() but this is more verbose."""
        
        return self.get_header_value()

    
    def get_last(self):
        """A method to return the node at the end of the linked list"""
        
        check_node=self.get_header_value()
                
        
        #The header is not empty
        if check_node is not None:
            
            #The original Node
            original_node=check_node
            
            #Check if the next link is None
            check_node=original_node.get_next_node()

            #Go through each node until the next node is none
            while check_node is not None:
                
                #The next node is not None, so set the header as the next node
                original_node=check_node
                
                #Check if the next node is None.
                check_node=original_node.get_next_node()
            
            return original_node
            
        #The header is empty to begin with. I.e. there is no existing nodes
        else:
            return None
        
    def remove_last(self):
        """A Function to unlink the last element
        
        We need to check two elements ahead to get the last node as the last one"""
        
        zeroth_node=self.get_header_value()
        
        #List is empty
        if zeroth_node is None:
            return
        
        #The zeroeth Node is not empty, so check the one-th element
        oneth_node=zeroth_node.get_next_node()
        
        #If the oneth element is empty, we need to remove the first node and empty the list
        if oneth_node is None:
            self.set_header_value(None)
            return

        #Two get next won't fail - Keey trying until None
        while zeroth_node.get_next_node().get_next_node() is not None:
            #zeroeth is the next
            zeroth_node = zeroth_node.get_next_node()
        
        #Two nodes from now is empty - Set Null
        zeroth_node.set_next_node(None)
        
        return

    
    ##------------END: EXTRA METHODS-----------------------###  
    
    
    
    
    ##------------BEGIN: MAGIC METHODS-----------------------###
    
    ###String Method
    def __str__(self):
        """A string method for the Linked list"""
        
        
        list_statement="Linked List is:"
        
        #The Header value is a node, i.e. there is a linked list
        if isinstance(self.get_header_value(),LinkedNode):
            current_node=self.get_header_value()
            
            #Go through each node until the link is null
            while current_node is not None:
                list_statement+="{}{}".format("\n",current_node)
                current_node=current_node.get_next_node()    
                
        #The linked list is empty
        else:
            list_statement+="{}{}".format("\n","(linked list is empty)")
        
        return """{}""".format(list_statement)
    
    ###Length Method.
    def __len__(self):
        """A length method for the Linked List - Number of links in the list"""
        
        length=0
                
        #The Header value is a node, i.e. there is a linked list
        if isinstance(self.get_header_value(),LinkedNode):
            
            current_node=self.get_header_value()

            #Go through each node until the link is null
            while current_node is not None:
                length+=len(current_node)
                current_node=current_node.get_next_node()    
                
        return length
    
    ###Equality Method
    #####NOTE: Linked Lists are equal only if all nodes are equal.
    def __eq__(self, other):
        """A method to check if two linked lists are equal"""
    
        equality=False
        
        
        self_current_node=self.get_header_value()
        other_current_node=other.get_header_value()      
        
        #While the other node equals the current node
        while self_current_node==other_current_node:
            
            #Check if the current node equals the other current node and is none. Both are at the end/
            if self_current_node is None:
                if other_current_node is None:
                    equality=True
                    break
                else:
                    break
                 
            #Self current node is not none - Break
            if other_current_node is None:
                break
                        
            self_current_node=self_current_node.get_next_node()
            other_current_node=other_current_node.get_next_node()
        
        return equality
    
    ###Representation Method - Note: Risk of Infinite Loop if add node ref: N1 - N2 - N3 - N1-...
    def __repr__(self):
        """Display representation"""
        
        return self.__str__()
    
    
    
    ##------------END: MAGIC METHODS-----------------------###
```

```python
class AltLinkedList:
    """A class for a linked list. A linked list is a series of linked nodes.
    
    This uses Alt Nodes"""
    
    __header=None
    
    ##------------BEGIN: INITIALISER-----------------------###
    def __init__(self,header=None):
        """Initialise a Linked Node"""
        self.set_header_value(header)
        
    ##------------END: INITIALISER-----------------------###
    
    
    ##------------BEGIN: HEADER VAL METHODS-----------------------###
    
    ###Header
    def set_header_value(self,header_value=None):
        """A setter for current node value"""
        
        if not isinstance(header_value,AltLinkedNode) and header_value is not None:
            print("Warning: You did not enter a linked node. You entered {}. Converting to node".format(type(header_value)))
            header_value=LinkedNode(header_value)
            
        self.__header=header_value
            
        return
        
    def get_header_value(self):
        """A getter for current node value"""
        return self.__header
    
    ##------------END: HEADER VAL METHODS-----------------------###
    
        
    ##------------BEGIN: ASSIGNMENT METHODS-----------------------###
    def add_first(self,new_header_node=None):
        """A method to add a new first element to the linked list"""
        
        header=self.get_header_value()
        
        #Checked that it's a new linked node, otherwise you shoudn't call it.
        if not isinstance(new_header_node, AltLinkedNode) and new_header_node is not None:
            print("Warning: You did not enter a node. You entered {}. Turning this into a node.".format(type(new_header_node)))
            new_header_node=AltLinkedNode(new_header_node)
            
        #If the Header Value is none, then the linked list was not defined so there is no existing header
        if header is None:

            #Set the header value to be the new node and done
            self.set_header_value(new_header_node)

        #The header value was not none, so it is a node.
        else:

            #Not adding a duplicate header
            if new_header_node!=header:
                #Set the new node to point to the old header node.
                new_header_node.set_next_node(header)

                #Set the header to be the new header node.
                self.set_header_value(new_header_node)

        return
    
    def add_last(self, new_tail_node=None):
        """A method to add a new node to the end of the linked list"""
        
        check_node=self.get_header_value()
        
        #Checked that it's a new linked node, otherwise you shoudn't call it.
        if not isinstance(new_tail_node, AltLinkedNode) and new_tail_node is not None:
            print("Warning: You did not enter a node. You entered {}. Turning this into a node.".format(type(new_tail_node)))
            new_tail_node=LinkedNode(new_tail_node)
        
        
        #The header is not empty
        if check_node is not None:
            
            #The original Node
            original_node=check_node
            
            #Check if the next link is None
            check_node=original_node.get_next_node()

            #Go through each node until the next node is none
            while check_node is not None:
                
                #The next node is not None, so set the header as the next node
                original_node=check_node
                
                #Check if the next node is None.
                check_node=original_node.get_next_node()

            
            #The header is now the tail as the next Node now returns None
            original_node.set_next_node(new_tail_node)
            
            #Set the New Node to have no Link as its the tail
            new_tail_node.set_next_node()
            
            return
            
        #The header is empty to begin with. I.e. there is no existing nodes
        else:
            self.set_header_value(new_tail_node)
        
        return
    
    def remove_first(self):
        """A method to remove the first element in the linked list"""
        
        #Get the next node from the header. 
        #--If it's not linked it's None so you're setting list to none.
        #--If it is linked it's a Node, so you're setting header to next node.
        
        header=self.get_header_value()
        
        if header is not None:
            self.set_header_value(header.get_next_node())
            
        return
    
    def traverse_list(self):
        """A method to traverse the list and print all elements"""

        print(self)
        
        return
    
    ##------------END: ASSIGNMENT METHODS-----------------------###
    
    
    ##------------BEGIN: EXTRA METHODS-----------------------###
    
    def get_first(self):
        """A method to get the first element of the linked list
        
        NOTEL Equivalent to get_header_value() but this is more verbose."""
        
        return self.get_header_value()

    
    def get_last(self):
        """A method to return the node at the end of the linked list"""
        
        check_node=self.get_header_value()
                
        
        #The header is not empty
        if check_node is not None:
            
            #The original Node
            original_node=check_node
            
            #Check if the next link is None
            check_node=original_node.get_next_node()

            #Go through each node until the next node is none
            while check_node is not None:
                
                #The next node is not None, so set the header as the next node
                original_node=check_node
                
                #Check if the next node is None.
                check_node=original_node.get_next_node()
            
            return original_node
            
        #The header is empty to begin with. I.e. there is no existing nodes
        else:
            return None
        
    def remove_last(self):
        """A Function to unlink the last element
        
        We need to check two elements ahead to get the last node as the last one"""
        
        zeroth_node=self.get_header_value()
        
        #List is empty
        if zeroth_node is None:
            return
        
        #The zeroeth Node is not empty, so check the one-th element
        oneth_node=zeroth_node.get_next_node()
        
        #If the oneth element is empty, we need to remove the first node and empty the list
        if oneth_node is None:
            self.set_header_value(None)
            return

        #Two get next won't fail - Keey trying until None
        while zeroth_node.get_next_node().get_next_node() is not None:
            #zeroeth is the next
            zeroth_node = zeroth_node.get_next_node()
        
        #Two nodes from now is empty - Set Null
        zeroth_node.set_next_node(None)
        
        return

    
    ##------------END: EXTRA METHODS-----------------------###  
    
    
    
    
    ##------------BEGIN: MAGIC METHODS-----------------------###
    
    ###String Method
    def __str__(self):
        """A string method for the Linked list"""
        
        
        list_statement="Linked List is:"
        
        #The Header value is a node, i.e. there is a linked list
        if isinstance(self.get_header_value(),AltLinkedNode):
            current_node=self.get_header_value()
            
            #Go through each node until the link is null
            while current_node is not None:
                list_statement+="{}{}".format("\n",current_node)
                current_node=current_node.get_next_node()    
                
        #The linked list is empty
        else:
            list_statement+="{}{}".format("\n","(linked list is empty)")
        
        return """{}""".format(list_statement)
    
    ###Length Method.
    def __len__(self):
        """A length method for the Linked List - Number of links in the list"""
        
        length=0
                
        #The Header value is a node, i.e. there is a linked list
        if isinstance(self.get_header_value(),AltLinkedNode):
            
            current_node=self.get_header_value()

            #Go through each node until the link is null
            while current_node is not None:
                length+=len(current_node)
                current_node=current_node.get_next_node()    
                
        return length
    
    ###Equality Method
    #####NOTE: Linked Lists are equal only if all nodes are equal.
    def __eq__(self, other):
        """A method to check if two linked lists are equal"""
    
        equality=False
        
        
        self_current_node=self.get_header_value()
        other_current_node=other.get_header_value()      
        
        #While the other node equals the current node
        while self_current_node==other_current_node:
            
            #Check if the current node equals the other current node and is none. Both are at the end/
            if self_current_node is None:
                if other_current_node is None:
                    equality=True
                    break
                else:
                    break
                 
            #Self current node is not none - Break
            if other_current_node is None:
                break
                        
            self_current_node=self_current_node.get_next_node()
            other_current_node=other_current_node.get_next_node()
        
        return equality
    
    ###Representation Method - Note: Risk of Infinite Loop if add node ref: N1 - N2 - N3 - N1-...
    def __repr__(self):
        """Display representation"""
        
        return self.__str__()
    
    
    
    ##------------END: MAGIC METHODS-----------------------###
```

# Analyse the o(n) methods. 

We will test the main methods of a linked list through the stack object. As both stacks and queues share these methods, results should be similar over operations which call the same underling LinkedList method.

As all methods which access the last element use the same logic in their methods, we test it using the push method for queues knowing the other methods will follow.

```python
stack_list={}

for test in range(10):
    result_dict={}
    
    sample_stack=MyStack()
    
    max_val=1000*test

    t1=time.time()
    for count in range(max_val):
        sample_stack.push(count)
        
    t2=time.time()
    time_delta=t2-t1
        
    result_dict['Object']='Stack'
    result_dict['Operation']='Push'
    result_dict['TimeToRun']=time_delta
    result_dict['Output']=sample_stack
    stack_list[max_val]=result_dict
    
```

```python
list_ops=list(stack_list.keys())
```

```python
y=[stack_list[x]['TimeToRun'] for x in list_ops]
```

# Plot: Time (s) vs Push Operations

We see that as anticipated the push operations have a linear time complexity. This will similarly apply for the print method and for methods which access the last element of a linked list (top, pop, remove_last) and enque for Queues. This validates our analysis.

```python
fig = plt.figure()
plt.plot(list_ops,y)
plt.show()
```
